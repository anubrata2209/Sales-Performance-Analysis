```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Data
#df = pd.read_csv(r"C:\Users\Anubrata Parida\MyPython\Projects\Sales Performance & Customer Insights Analysis\sales_data.csv")
df = pd.read_csv("sales_data.csv")
df.head()
```

```python
#Data Cleaning
# Check missing values
df.isnull().sum()
```

```python
# Convert date
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Save cleaned data
df.to_csv("sales_cleaned.csv", index=False)
```

```python
#Exploratory Data Analysis (EDA)

sales_trend = df.groupby('Order_Date')['Sales'].sum()

plt.figure(figsize=(10,5))
sales_trend.plot()
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
#Sales by Category

plt.figure(figsize=(7,5))
sns.barplot(x='Category', y='Sales', data=df)
plt.title("Sales by Category")
plt.show()
```

```python
#Profit Distribution

sns.histplot(df['Profit'], bins=30, kde=True)
plt.title("Profit Distribution")
plt.show()
```

```python
#Top 10 Customers
top_customers = df.groupby('Customer_Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_customers
```
